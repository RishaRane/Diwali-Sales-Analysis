# Exploratory Data Analysis

### Gender

1. countplot

```python
diwali_sales.Gender.unique()
```

```python
# Setting the background as default parameters
sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 12
```

Way 1 :

```python
plt.title('Sales distribution w.r.t Gender');
colors = ['hotpink', 'steelblue']
sns.countplot( x = 'Gender', data = diwali_sales, palette = colors);
```

To get the total sales count as a label we run a for loop

Way 2:

```python
gender_countplot = sns.countplot( x = 'Gender', data = diwali_sales);
for bars in gender_countplot.containers:
    gender_countplot.bar_label(bars)
```

Way 3:

```python
# Without as_index = False i.e Gender will be the index
diwali_sales.groupby(['Gender'])['Amount'].sum()
```

```python
# With as_index = False i.e Gender will be the index
amount_group = diwali_sales.groupby(['Gender'], as_index = False)['Amount'].sum()
amount_group
```

2. barplot

```python
colors = ['hotpink', 'steelblue'];
sns.barplot( x = 'Gender', y = 'Amount', data = amount_group, palette = colors);
```

Conclusion : The maximum purchases are by 'Females'

# State

```python
# By 'Amount' we mean that total amount spent by a people of a state on shopping in diwali
state_sale = diwali_sales.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
state_sale
```

1. Line chart

```python
# Total amount spent by people of x state
plt.title('Total amount spent by people w.r.t State')
plt.plot(state_sale.State, state_sale.Amount);
plt.xticks(rotation = 90);
```

2. Barplot

```python
# Total amount spent by people of x state
plt.title('Total amount sales by people w.r.t State')
sns.barplot(x = 'State', y = 'Amount', data = state_sale);
plt.xticks(rotation = 90);
```

```python
plt.title('Total amount sales from top 5 states')
state_sale_top_10 = diwali_sales.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(5)
amount_sales = sns.barplot(x = 'State', y = 'Amount', data = state_sale_top_10);
plt.xticks(rotation = 90);
```

```python
# Total orders by people of x state
state_order = diwali_sales.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False)
state_order
```

```python
plt.title('Total Orders placed w.r.t State')
order_plot = sns.barplot(x = 'State', y = 'Orders', data = state_order);
plt.xticks(rotation = 90);
for bars in order_plot.containers:
    order_plot.bar_label(bars)
```

3. Countplot

```python
#Total people from x state w.r.t Gender
plt.title('Total no. of people shopping w.r.t State');
sns.countplot( x = 'State', hue = 'Gender', data = diwali_sales);
plt.xticks(rotation = 90);
```

Conclusion : By this we can conclude that maximum orders are placed by people of 'Uttar Pradesh'. To be more specific from U.P total of 4807 people placed orders

```python
plt.title('Total No. of order from top 5 states')
state_order_top_10 = diwali_sales.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(5)
order_top_10 = sns.barplot(x = 'State', y = 'Orders', data = state_order_top_10);
plt.xticks(rotation = 90);
for bars in order_top_10.containers:
    order_top_10.bar_label(bars)
```

# Marital Status

```python
diwali_sales.columns
```

```python
plt.title('Sales w.r.t Marital Status')
marry_count = sns.countplot(x = 'Marital_Status', data = diwali_sales);
for bars in marry_count.containers:
    marry_count.bar_label(bars)
```

'0' refers to married people while '1' refers to unmarried people

Conclusion : Married people have done more shopping

```python
ms_gender = diwali_sales.groupby(['Marital_Status', 'Gender'])['Amount'].sum()
ms_gender
```

```python
ms_gender = diwali_sales.groupby(['Marital_Status', 'Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
ms_gender
```

```python
plt.title('Sales w.r.t Marital Status (Sex-wise Distribution)')
sns.barplot(x = 'Marital_Status', y = 'Amount', hue = 'Gender', data = ms_gender);
```

Conclusion : Married Woman did more shopping than married men and unmarried woman

# Product Category

```python
diwali_sales.columns
```

```python
diwali_sales.Product_Category.unique()
```

1. Barplot

```python
product_sale = diwali_sales.groupby(['Product_Category'], as_index = False)['Amount'].count().sort_values(by = 'Amount', ascending = False).head(10)
product_sale
```

```python
plt.figure(figsize = (21, 5))
plt.title('Topmost 10 Products purchased by Customers')
product_count = sns.barplot(x = 'Product_Category', y = 'Amount', data = product_sale);
plt.plot(product_sale.Product_Category, product_sale.Amount, 'o--r');
plt.ylabel('Count of Products');
for bars in product_count.containers:
    product_count.bar_label(bars)
```

Conclusion : Productly that most people shopped for was 'Clothing & Apprael' followed by Food. To be more specific total 2655 clothing items were sold while total 2490 food items
